# PLOTS

# 1.Has the average incident resolution time decreased over the years for each country?

```python


f_obj=sns.FacetGrid(df,col='Country',col_wrap=4,height=4,sharey=False)
f_obj.figure.suptitle('Year-wise Average Incident Resolution Time in Each Country')

# for ax in f_obj.axes.flatten():
#  for label in ax.get_xticklabels():
#         label.set_rotation(45)

for ax in f_obj.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)        

f_obj.map_dataframe(sns.barplot,'Year','Incident Resolution Time (in Hours)')


plt.subplots_adjust(hspace=0.5)
plt.show()
plt.tight_layout()

```

# Observation:
1. No, average incident resolution time has not decreased over the years for each country
2. Their is variation in the incident resolution time
3. And one more thing we can observe that ,their is no constant improvement in incident resolution time

# 2.Whether their is an improvement in an Average Incident Resolution Time over years?

```python
a=df.groupby(['Year'])['Incident Resolution Time (in Hours)'].mean().sort_index(ascending=False).reset_index()
sns.lineplot(data=a,x='Year',y='Incident Resolution Time (in Hours)')
```

# Observation:
1. Their is no improvement in Incident Resolution Time over years

# 3. Has the number of cyber attacks increased or decreased year by year in each country

```python
attack_count=df.groupby(['Year','Country']).size().reset_index()
attack_count.columns=['Year','Country','attack_counts']
attack_count
```

```python

g= sns.FacetGrid(attack_count, col='Country', col_wrap=4, height=4, sharey=False)
g.figure.suptitle('Year-wise Number of Cyber Attacks in Each Country')
g.map_dataframe(sns.barplot, x='Year', y='attack_counts')


for ax in g.axes.flatten():
 for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.subplots_adjust(hspace=0.4)
plt.show()
plt.tight_layout()
```

# Observation:
1. No,Their is variation in the number of cyber attacks , year by year in each country
  i.e either it is not completely increased neither not completely decreased

# 5.Which countries face the most attacks by hacker groups or nation states?

```python
newdf=df.loc[df['Attack Source'].isin(['Hacker Group','Nation-state']),::]
a=sns.countplot(data=newdf,x='Country',hue='Attack Source')
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=90)  
plt.title('Count of Attacks from Hacker group or Nation-State in each Country')
```

# 9.Is there a relation between number of affected users and financial loss?
